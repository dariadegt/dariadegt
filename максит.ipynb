```python
import random as rd

S = [0, 0]
#суммы игроков
P = 0
#счетчик игрока. ходы

#def the initial Tab
a = list("000")
for i in range(0,3):
    a[i] = rd.randint(1,9)
    

b = list("000")
for i in range (0,3):
    b[i] = rd.randint(1,9)
   

c = list("000")
for i in range (0,3):
    c[i] = rd.randint(1,9)

EndPl = 0

#The main func for players
def PL():
   print (" ", "0", "1", "2" )
   print("--------")
   print ("a|", *a)
   print ("b|", *b)
   print ("c|", *c)
   

   print("ходит игрок №" , P+1)
   row = input("введите ряд:")
   col = int(input("введите столбец:"))

   if row == "a":
     S[P] += a[col]
     a[col]="x"
     rowpl1 = a

   if row == "b":
      S[P] += b[col]
      b[col] = "x"
      rowpl1 = b

   if row =="c":
     S[P] += c[col]
     c[col] = "x"
     rowpl1 = c
#end of data entering

   if P==0:
     colpl2 = [a[col], b[col], c[col]]
     d2 = colpl2.count("x")
#проверка наличия хода
     if d2 == 3:
         dif = 0
         print("у противника нет хода")
     if d2 != 3:
         dif = 1
     pass  
   if P == 1 :
        d1 = rowpl1.count("x")
        if d1 == 3:
            dif = 0
            print("у противника нет хода")
        if d1 != 3:
            dif = 1
            print("продолжайте")
            pass
   
   print("Сумма очков игрока №1 =", S[0])
   print("Сумма очков игрока №2 =", S[1])

   colpl2 = [a[col], b[col], c[col]]
   d2 = colpl2.count("x")
   d1 = rowpl1.count("x")
   # d1 d2 координаты 

   return(dif,d1,d2)


#проверка на возможность хода(если в строке и столбце есть х(типо на пересечении), не позволяющие больше сделать ход)

while EndPl < 6:
      PL_1 = PL()
      dif1 = PL_1[0]
      EndPl = PL_1[1] + PL_1[2]
      P = (P+dif1)%2
      if EndPl == 6:
        if S[0] > S[1]:
          print("победил игрок №1")
        else:
          print("победил игрок №2")
        print("счет победившего игрока =", max([S[0],S[1]]))






  

```
